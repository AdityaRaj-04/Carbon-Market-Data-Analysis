```python
import pandas as pd
```

```python
carban_market = pd.read_csv('GSF Registry Credits.csv', encoding='ISO-8859-1')
```

```python
carban_market.head()
```

```python
carban_market.isnull().sum()
```

```python
carban_market['Methodology'].fillna('Unknown', inplace=True)
```

```python
carban_market.drop(columns=['POA GSID'], inplace=True)
```

```python
carban_market.dropna(subset=['Retirement Date'], inplace=True)
```

```python
carban_market.drop(columns=['Note'], inplace=True)
```

```python
carban_market['Aeroplane Operator Name'].fillna('Unknown', inplace=True)
```

```python
print(carban_market.isnull().sum())
```

```python
carban_market['Issuance Date'] = pd.to_datetime(carban_market['Issuance Date'], errors='coerce')
carban_market['Retirement Date'] = pd.to_datetime(carban_market['Retirement Date'], errors='coerce')

# Extract year
carban_market['Issuance Year'] = carban_market['Issuance Date'].dt.year
carban_market['Retirement Year'] = carban_market['Retirement Date'].dt.year

# Total issuances per year
total_issuances_per_year = carban_market.groupby('Issuance Year')['Quantity'].sum()

# Total retirements per year
total_retirements_per_year = carban_market.groupby('Retirement Year')['Quantity'].sum()
```

```python
top_project_types = carban_market.groupby('Project Type')['Quantity'].sum().sort_values(ascending=False)
```

```python
geo_distribution = carban_market['Country'].value_counts()
```

```python
import matplotlib.pyplot as plt

# Plot total issuances per year
plt.figure(figsize=(10, 6))
total_issuances_per_year.plot(kind='bar')
plt.title('Total Issuances per Year')
plt.xlabel('Year')
plt.ylabel('Total Issuances')
plt.show()
```

```python
# Plot total retirements per year
plt.figure(figsize=(10, 6))
total_retirements_per_year.plot(kind='bar')
plt.title('Total Retirements per Year')
plt.xlabel('Year')
plt.ylabel('Total Retirements')
plt.show()
```

```python
# Plot top project types
plt.figure(figsize=(10, 6))
top_project_types.plot(kind='bar')
plt.title('Top Project Types by Issuance Volume')
plt.xlabel('Project Type')
plt.ylabel('Total Issuances')
plt.show()
```

```python
# Plot geographical distribution
plt.figure(figsize=(12, 8))
geo_distribution.plot(kind='bar')
plt.title('Geographical Distribution of Projects')
plt.xlabel('Country')
plt.ylabel('Number of Projects')
plt.show()
```

Top Project Developer Profiling

```python
# Group by 'Project Developer' and sum 'Quantity' to get total issuance volume
top_developers = carban_market.groupby('Project Developer')['Quantity'].sum().sort_values(ascending=False).head(30)

# Create a DataFrame for top developers
top_developers_df = top_developers.reset_index()
print(top_developers_df.head())
```

```python
# Create a DataFrame for top developers with additional columns
top_developers_df = pd.DataFrame({
    'Developer Name': top_developers_df['Project Developer'],
    'Website URL': [''] * len(top_developers_df),  # Fill this in manually or via scraping
    'Main Project Types': [''] * len(top_developers_df),  # Fill this in manually
    'Associated Projects': [''] * len(top_developers_df),  # Fill this in manually
    'Total Credits Issued': top_developers_df['Quantity'],
    'Total Credits Retired': [''] * len(top_developers_df),  # Fill this in if available
    'Total Credits Remaining': [''] * len(top_developers_df)  # Calculate if needed
})
```

Credit Buyer Analysis

```python
# Extract a list of buyers
credit_buyers = carban_market.groupby('Project Name')['Quantity'].sum().reset_index()

# Create a DataFrame for credit buyers
credit_buyers_df = credit_buyers.rename(columns={'Project Name': 'Buyer Name', 'Quantity': 'Total Credits Retired'})
print(credit_buyers_df.head())
```

```python
# Example analysis of top credit buyers
top_buyers = credit_buyers_df.sort_values(by='Total Credits Retired', ascending=False).head(10)
print(top_buyers)
```

```python
# Save cleaned and analyzed carbon project data
carban_market.to_csv('cleaned_carbon_project_data.csv', index=False)

# Save top project developers profile
top_developers_df.to_csv('top_project_developers_profile.csv', index=False)

# Save credit buyers data
credit_buyers_df.to_csv('carbon_credit_buyers.csv', index=False)
```
